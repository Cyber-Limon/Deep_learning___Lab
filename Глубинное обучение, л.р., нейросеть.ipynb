## Нейросеть

### Импорт библиотек:
- "pandas"     - загрузка данных из Excel-файла в массивы составов и свойств;
- "openpyxl"   - чтение ".xlsx" файлов;
- "sklearn"    - нормализация данных и разделение выборки на обучающую и тестовую;
- "tensorflow" - создание и обучение нейросети для прогнозирования свойств.

```python
import numpy  as np
import pandas as pd

from tensorflow              import keras
from sklearn.preprocessing   import StandardScaler
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
```

### Загрузка данных:
- "compound"   - массив составов сплавов;
- "properties" - массив свойств сплавов.

```python
compound   = pd.read_excel("Глубинное обучение, л.р., датасет.xlsx", usecols="B:V")
compound   = compound.values.tolist()
compound   = np.array(compound)

properties = pd.read_excel("Глубинное обучение, л.р., датасет.xlsx", usecols="W:AD")
properties = properties.values.tolist()
properties = np.array(properties)
```

### Разделение данных на тренировочную и тестовую выборки:
- "X_train" - тренировочные составы;
- "X_test"  - тестовые составы;
- "y_train" - тренировочные свойства;
- "y_test"  - тестовые свойства.

```python
X_train, X_test, y_train, y_test = train_test_split(
    compound,
    properties,
    test_size=0.2,
    random_state=1107,
    shuffle=True
)
```

### Нормализация данных

```python
scaler_X = StandardScaler()
X_train  = scaler_X.fit_transform(X_train)
X_test   = scaler_X.transform(X_test)

scaler_Y = StandardScaler()
y_train  = scaler_Y.fit_transform(y_train)
y_test   = scaler_Y.transform(y_test)
```

### Архитектура модели

```python
model = keras.Sequential([
    keras.Input(shape=(X_train.shape[1],)),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.4),
    Dense(64,  activation='relu'),
    Dropout(0.5),
    Dense(32,  activation='relu'),
    Dropout(0.6),
    Dense(y_train.shape[1])
])

model.summary()
```

### Обучение модели

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

log = model.fit(
    X_train, y_train,
    batch_size=16,
    epochs=10,
    validation_split=0.2,
    verbose=1
)
```

### Тестирование модели

```python
model.evaluate(X_test, y_test, verbose=1)
```

## Графический интерфейс пользователя (GUI)

### Импорт библиотек:
- "tkinter" - создание графического интерфейса для ввода состава и вывода прогноза.

```python
from tkinter import *
```

### Названия ячеек

```python
compound   = pd.read_excel("Глубинное обучение, л.р., датасет.xlsx", nrows=0, usecols="B:V")
compound   = compound.columns.tolist()

properties = pd.read_excel("Глубинное обучение, л.р., датасет.xlsx", nrows=0, usecols="W:AD")
properties = properties.columns.tolist()
```

### Обработка событий:
- "clean_compound()"     - очищает состав и строку предупреждения;
- "clean_properties()"   - очищает свойства и строку предупреждения;
- "on_compound_change()" - очищает строку предупреждения;
- "forecast()"           - выполняет прогноз;
- "validate_input()"     - обрабатывает вводимый текст.

```python
def clean_compound():
    clean_properties()

    for entry in entries_compound:
        entry.delete(0, END)



def clean_properties():
    title_3.config(text="")

    for entry in entries_properties:
        entry.config(state='normal')
        entry.delete(0, END)
        entry.config(state='readonly')



def on_compound_change(event):
    title_3.config(text="")

    clean_properties()



def forecast():
    current_compound = []
    for entry in entries_compound:
        current_compound.append(float(entry.get()) if entry.get() != '' else 0)

    if sum(current_compound) != 100:
        title_3.config(text=f"ПРЕДУПРЕЖДЕНИЕ: неправильный ввод - сумма элементов состава ({sum(current_compound)}) не равна 100")
        return
    else:
        title_3.config(text="")

    current_compound   = scaler_X.transform([current_compound])
    current_properties = model.predict(np.array(current_compound))
    current_properties = scaler_Y.inverse_transform(current_properties)

    for i in range(len(entries_properties)):
        entries_properties[i].config(state='normal')
        entries_properties[i].delete(0, END)
        entries_properties[i].insert(0, round(current_properties[0][i], 5))
        entries_properties[i].config(state='readonly')



def validate_input(char, current_value):
    if char == '' or char.isdigit():
        return True

    if char == '.' and '.' not in current_value:
        return True

    return False
```

### Интерфейс

```python
root = Tk()
root.title("Прогнозирование свойств")



columns = 16
font = ('Calibri Bold', 20)

vcmd = (root.register(validate_input), '%S', '%s')



title_1 = Label(root, text="ПРОГНОЗИРОВАНИЕ СВОЙСТВ", font=('Calibri Bold', 25))
title_1.grid(row=0, column=0, columnspan=columns, pady=5)

title_2 = Label(root, text="Введите процентный состав", font=font)
title_2.grid(row=2, column=0, columnspan=columns, pady=(30, 0))

entries_compound = []
for i in range(21):
    r = i // 8 + 3
    c = (i * 2) % columns

    label = Label(root, width=5, text=compound[i], font=font)
    label.grid(row=r, column=c, pady=10)

    entry = Entry(root, width=5, font=font, justify='center', validate='key', validatecommand=vcmd)
    entry.grid(row=r, column=(c + 1), padx=10)
    entry.bind('<Key>', on_compound_change)
    entries_compound.append(entry)

button_clean = Button(root, text="Очистить состав", font=font, cursor='hand2', bg='#ff4444', fg='white', command=clean_compound)
button_clean.grid(row=5, column=10, columnspan=3, sticky='ew', padx=10)

button_forecast = Button(root, text="Выполнить прогноз", font=font, cursor='hand2', bg='#44ff44', fg='white', command=forecast)
button_forecast.grid(row=5, column=13, columnspan=3, sticky='ew', padx=10)

title_3 = Label(root, text="", font=font, fg="red")
title_3.grid(row=6, column=0, columnspan=columns)

title_4 = Label(root, text="Прогноз", font=font)
title_4.grid(row=7, column=0, columnspan=columns)

entries_properties = []
for i in range(8):
    r = i // 4 + 8
    c = (i * 4) % columns

    label = Label(root, width=10, text=properties[i], font=font)
    label.grid(row=r, column=c, columnspan=2, pady=10)

    entry = Entry(root, width=10, font=font, justify='center', state='readonly')
    entry.grid(row=r, column=(c + 2), columnspan=2, padx=10)
    entries_properties.append(entry)



root.mainloop()
```
